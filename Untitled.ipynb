```python
import pandas as pd 
import numpy as np 

df= pd.read_csv('sattelite launches.csv')
```

```python
df.head(30)
```

```python
df.columns
df_gantinama = df.rename(columns={'norad_id':'Nomor ID','cospar_id':'Kode Satelit','name':'Nama Satelit','launch_date':'Tanggal Luncur',
                  'flight_ended':'Tanggal Berakhir','status':'Status','destination':'Destinasi','owner':'Owner'})
df_gantinama.head(30)
```

```python
df_gantinama.isna()
df_gantinama.isnull().sum()
df_gantinama['Tanggal Berakhir'].fillna('-',inplace=True)
df_gantinama['Status'].fillna('Unknown',inplace=True)
df_gantinama['Destinasi'].fillna('-',inplace=True)
df_gantinama.head(30)
```

```python
df_gantinama['Tanggal Luncur']
```

```python
df_gantinama['Bulan Luncur'] = x[1]
df_gantinama['Tahun Luncur'] = x[2]

df_gantinama.head(30)
```
